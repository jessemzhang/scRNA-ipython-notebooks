- Isolate a batch i (after getting rid of controls and low-coverage cells)
- Filter out genes that:
    - are expressed in <= thresh1 % of the cells AND have max expression less than thresh2
    - have large kurtosis (>= 2.9)
- Estimate R
- Visualize R (PCA), and label the D1D2 genes
    - Try isolating the columns of R that correspond to D1D2 genes
- Repeat 4 with abs(R)
- Repeat 1-5 with all batches

```python
# Load data, packages

import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

X = np.loadtxt('/data/jessez/Gene_count_datasets/Datasets/batch_all/X')
genes = np.loadtxt('/data/jessez/Gene_count_datasets/Genes/genes.txt',dtype=str)
genes_d1 = ['Isl1','Drd1','Sfxn1','Nrxn1']
genes_d2 = ['Drd2','Penk','Sp9','Gpr52','Gpr88']
batches = np.loadtxt('/data/jessez/Gene_count_datasets/Batch_nums/batch_num_all.txt')
M,N = np.shape(X)
```

```python
# Isolate batch
batch = 24
X_batch = X[batches == batch]
M_batch = np.sum(batches == batch)

# Filter out genes that:
# are expressed in <= thresh1 % of the cells
# have mean expression (in non-zero entries) less than thresh2
# have high kurtosis
save_ind = np.zeros(N)+1
k = kurtosis(X_batch,axis=0) 
thresh1 = 0.3
thresh2 = np.median(X_batch[X_batch > 0])
for i in range(N):
    x = X_batch[:,i]
    if np.sum(x>0)/float(np.size(x)) < thresh1 and np.mean(x[x>0]) < thresh2 and k[i] > -0.1:                             
        save_ind[i] = 0
save_ind = np.where(save_ind.astype('int'))
X_batch = np.squeeze(X_batch[:,save_ind])
genes_batch = genes[save_ind]

# Estimate R
R = np.corrcoef(np.transpose(X_batch))
```

```python
# Run this if you want to look at R for all batches, not just for individual batches
R = np.corrcoef(np.transpose(X))
genes_batch = genes
np.shape(genes_batch)
```

```python
# Visualize R using PCA 

# Threshold R to make it more sparse (easier for computation)
R_abs = np.abs(R)
R_thresh = R.copy()
R_thresh[R_abs < .15] = 0

U,S,V = sp.sparse.linalg.svds(R_thresh,k=2)

# Plot genes
%matplotlib inline
plt.figure(figsize=(15,10))
plt.scatter(U[:,0],U[:,1])
for gene in genes_d1:
    ind = np.where(genes_batch == gene)
    plt.scatter(U[ind,0],U[ind,1],255,c=[1,0,0])
for gene in genes_d2:
    ind = np.where(genes_batch == gene)
    plt.scatter(U[ind,0],U[ind,1],255,c=[0,1,0])
plt.show()
```

```python
# Repeat above using abs(R)

# Threshold R to make it more sparse (easier for computation)
R_abs = np.abs(R)
R_thresh = R_abs.copy()
R_thresh[R_abs < .15] = 0
np.min(R_thresh)

U,S,V = sp.sparse.linalg.svds(R_thresh,k=2)

# Plot genes
%matplotlib inline
plt.figure(figsize=(15,10))
plt.scatter(U[:,0],U[:,1])
for gene in genes_d1:
    ind = np.where(genes_batch == gene)
    plt.scatter(U[ind,0],U[ind,1],255,c=[1,0,0])
for gene in genes_d2:
    ind = np.where(genes_batch == gene)
    plt.scatter(U[ind,0],U[ind,1],255,c=[0,1,0])
plt.show()
```

```python
# This code changes R to R for just d1d2 genes

save_ind = np.zeros(np.size(genes_batch))

for i in range(np.size(genes_batch)):
    if genes_batch[i] in genes_d1 or genes[i] in genes_d2:
        save_ind[i] = 1
save_ind = np.where(save_ind.astype('int'))

R_d1d2 = np.squeeze(R[:,save_ind])
R_d1d2 = R_d1d2[save_ind]
```
