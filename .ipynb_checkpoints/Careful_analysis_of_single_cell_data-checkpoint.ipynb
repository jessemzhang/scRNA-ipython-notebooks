The purpose of this iPython notebook is discover some structures in the single cell data.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Functions for plotting interesting information:

```python
# Plot (proportion of zeros for a gene) vs. (mean in non-zeros)
def plot_proportion_zeros_to_mean_in_nonzeros(X):
    # For each gene, compute the number of cells that don't express this each
    proportion_zeros = np.sum(X == 0,axis=0)/float(np.shape(X)[0])
    # For each gene, compute the mean of non-zero cells
    mean_in_nonzeros = []
    for i in range(np.shape(X)[1]):
        x = X[:,i]
        mean_in_nonzeros.append(np.mean(x[x > 0]))
    # Plot stuff
    plt.figure(figsize=(15,5))
    plt.scatter(mean_in_nonzeros,proportion_zeros)
    plt.xlabel('mean in non-zero cells')
    plt.ylabel('proportion of cells that have 0')
    plt.show()

# Plot the histogram of a gene
def plot_hist_of_gene(X,gene,genes):
    gene_ind = np.where(genes == gene)
    X_gene = np.squeeze(X[:,gene_ind])
    plt.figure(figsize=(15,5))
    plt.hist(X_gene,bins=np.size(X_gene)/20)
    plt.title(gene)
    plt.show()
    
# Plot a vector
def plot_vect(x):
    plt.figure(figsize=(15,5))
    plt.plot(range(np.size(x)),x)
    plt.show()
```
