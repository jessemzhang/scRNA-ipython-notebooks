```python
# Load data

import numpy as np
X = np.transpose(np.loadtxt('/data/jessez/Gene_count_datasets/Datasets/batch_all/X_Log_DESeq'))
batch_num = np.loadtxt('/data/jessez/Gene_count_datasets/Batch_nums/batch_num_all.txt')
```

```python
from sklearn.decomposition import NMF
import matplotlib.pyplot as plt

def test_batch(batch,batch_num,X):
    X_temp = X[:,batch_num==batch]
    nmf = NMF(n_components=2, random_state=1).fit(X_temp)
    X_NMF = nmf.components_
    plt.scatter(X_NMF[0], X_NMF[1])    
```

```python
# Run NMF

plt.figure(figsize=(15,30))

ii = 1
for batch in np.unique(batch_num): 
    plt.subplot(7,1,ii)
    plt.title(str(batch) + ', which has ' + str(np.sum(batch_num == batch)) + ' samples')
    test_batch(batch,batch_num,X)
    ii += 1
    
plt.show()
```
