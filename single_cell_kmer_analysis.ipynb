```python
# Define functions, import libraries here
import sys
import os
import numpy as np
from scipy.sparse import *

# Compute distance according to Valiant
def valiant_dist(i,j,X):
    x = X[i].astype(float)
    y = X[j].astype(float)
    a = x + y
    b = x - y
    c = (a != 0)
    # for entries that have x + y = 0, the contribution should be -1)                                         
    return (np.sum((np.square(b[c])-a[c])/a[c])-(len(x)-np.sum(c)))/len(x)

# Compute total variation between two distributions
def total_variation(i,j,X):
    x = X[i].astype(float)
    y = X[j].astype(float)
    return np.sum(np.absolute(x-y))

# Map kmer to index
def k_to_i(kmer):    
   alphabet = {'A':0, 'C':1, 'G':2, 'T':3}
   index = map(lambda x:alphabet[x], kmer)
   index = int(''.join(map(str,index)),4)
   return index

# Load data from kmer file (file with three columns: 
def load_k(filename,k):
   f = open(filename)
   x = np.zeros(4 ** k) # dictionary to store relative kmer abundances       
   for line in f:
       temp = line.split()
       x[k_to_i(temp[0])] = temp[2]
   f.close() 
   return x
```

```python
# Load data into matrix (X will have N samples, and each sample will be described with a length-4^k vector
%timeit
X = []
a = os.listdir('/data/jessez/10_mers_distributions_no_N')
a.sort(key=lambda x:int(x.split('_')[1]))
for file in a:
    X.append(load_k('/data/jessez/10_mers_distributions_no_N/' + file,10))
    print file + ' loaded..'
X = np.array(X)
```

```python
# Functions for performing PCA, t-SNE, and kmeans

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn import manifold
from sklearn import cluster

# obtain PCs and plot by batch number
def jz_PCA(X):
    X_zm = X
    X_bar = np.mean(X,axis=0)
    U, s, V = np.linalg.svd(np.dot(X_zm,np.transpose(X_zm)), full_matrices=True)
    return U

# obtain two major directions from t-SNE and plot by batch number
def jz_tSNE(X):
    tsne = manifold.TSNE(n_components=2, random_state=0);
    X_tsne = tsne.fit_transform(X);
    return X_tsne

# obtain labels via kmeans
def jz_kmeans(X):
    k_means = cluster.KMeans(n_clusters=2)
    k_means.fit(X)
    labels = k_means.labels_
    return labels
```

```python
# PCA on original matrix
%matplotlib inline
batch_num = np.loadtxt('/data/jessez/Gene_count_datasets/Batch_nums/batch_num_all.txt')
X_pca = jz_PCA(X)

plt.figure(figsize=(15,15))

# plot and color points using batch numbers
batches = np.unique(batch_num)
plt.subplot(2,1,1)
for batch in batches:
    plt.scatter(X_pca[batch_num == batch,0], X_pca[batch_num == batch, 1],c=np.random.rand(3))
    
# plot and color using kmeans labels
pca_labels = jz_kmeans(X_pca)
plt.subplot(2,1,2)
plt.scatter(X_pca[pca_labels == 0,0],X_pca[pca_labels == 0,1],c=np.random.rand(3))
plt.scatter(X_pca[pca_labels == 1,0],X_pca[pca_labels == 1,1],c=np.random.rand(3))

plt.show()
```

```python
# PCA on individual batches
%matplotlib inline
batch_num = np.loadtxt('/data/jessez/Gene_count_datasets/Batch_nums/batch_num_all.txt')
batches = np.unique(batch_num)

plt.figure(figsize=(15,15))
subplot_ind = 1
for batch in batches:
    Y = jz_PCA(X[batch_num == batch])
    plt.subplot(4, 2, subplot_ind)
    plt.scatter(Y[:,0], Y[:, 1])
    plt.title(batch)
    subplot_ind += 1
plt.savefig('/home/jessez/batch_10mer_pca.png')
plt.show()
```
